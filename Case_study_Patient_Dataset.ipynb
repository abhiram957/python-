```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df =pd.read_csv('heart.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
df.describe()
```

```python
df.target.value_counts()
```

```python
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
sns.countplot(x=df.target,hue='sex',data=df,)
plt.xticks([0,1],labels=['Not Diseases ','Disease'])
plt.title("Target Distribution");
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA538','red'])
plt.legend(["Haven't Disease","Have Disease"])
plt.title("Heart Diseases Freaquency for Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='blue',label='Not Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='red',label='Have Disease')
plt.legend()
plt.title('Heart Disease Distribution for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
